```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
fgd5 = pd.read_csv("~/count_matrix/fgd5", sep='\t', header=0)
emx2os = pd.read_csv("~/count_matrix/emx2os", sep='\t', header=0)
ac = pd.read_csv("~/count_matrix/ac", sep='\t', header=0)
rp11 = pd.read_csv("~/count_matrix/rp11", sep='\t', header=0)
ctd = pd.read_csv("~/count_matrix/ctd", sep='\t', header=0)
jpx = pd.read_csv("~/count_matrix/jpx", sep='\t', header=0)
mapkapk_h3k4me3 = pd.read_csv("~/count_matrix/mapkapk_h3k4me3", sep='\t', header=0)
mapkapk_h3k27me3 = pd.read_csv("~/count_matrix/mapkapk_h3k27me3", sep='\t', header=0)
```

```python
def make_table (count_matrix):
    max_nc_01 = max(count_matrix.nc_01)
    max_nc_02 = max(count_matrix.nc_02)
    max_kd_01 = max(count_matrix.kd_01)
    max_kd_02 = max(count_matrix.kd_02)
    max_nc_01_input = max(count_matrix.nc_01_input)
    max_nc_02_input = max(count_matrix.nc_02_input)
    max_kd_01_input = max(count_matrix.kd_01_input)
    max_kd_02_input = max(count_matrix.kd_02_input)
    max_read_num = max(max_nc_01, max_nc_02, max_kd_01, max_kd_02, max_nc_01_input, max_nc_02_input, max_kd_01_input, max_kd_02_input)
    
    nc_01 = pd.crosstab(index=count_matrix['nc_01'], columns='count')
    nc_01 = pd.DataFrame({"nc_01":nc_01['count']})
    nc_02 = pd.crosstab(index=count_matrix['nc_02'], columns='count')
    nc_02 = pd.DataFrame({"nc_02":nc_02['count']})
    kd_01 = pd.crosstab(index=count_matrix['kd_01'], columns='count')
    kd_01 = pd.DataFrame({"kd_01":kd_01['count']})
    kd_02 = pd.crosstab(index=count_matrix['kd_02'], columns='count')
    kd_02 = pd.DataFrame({"kd_02":kd_02['count']})
    nc_01_input = pd.crosstab(index=count_matrix['nc_01_input'], columns='count')
    nc_01_input = pd.DataFrame({"nc_01_input":nc_01_input['count']})
    nc_02_input = pd.crosstab(index=count_matrix['nc_02_input'], columns='count')
    nc_02_input = pd.DataFrame({"nc_02_input":nc_02_input['count']})
    kd_01_input = pd.crosstab(index=count_matrix['kd_01_input'], columns='count')
    kd_01_input = pd.DataFrame({"kd_01_input":kd_01_input['count']})
    kd_02_input = pd.crosstab(index=count_matrix['kd_02_input'], columns='count')
    kd_02_input = pd.DataFrame({"kd_02_input":kd_02_input['count']})
    
    read_number = [i for i in range(0, max_read_num+1, 1)]
    
    table = pd.DataFrame({"read_number":read_number,})
    table['nc_01'] = 0
    table['nc_02'] = 0
    table['kd_01'] = 0
    table['kd_02'] = 0
    table['nc_01_input'] = 0
    table['nc_02_input'] = 0
    table['kd_01_input'] = 0
    table['kd_02_input'] = 0
    
    sequence = nc_01.index.values
    for i in sequence:
        table.loc[i, 'nc_01'] = nc_01.loc[i].values[0]
        
    sequence = nc_02.index.values
    for i in sequence:
        table.loc[i, 'nc_02'] = nc_02.loc[i].values[0]
        
    sequence = kd_01.index.values
    for i in sequence:
        table.loc[i, 'kd_01'] = kd_01.loc[i].values[0]
        
    sequence = kd_02.index.values
    for i in sequence:
        table.loc[i, 'kd_02'] = kd_02.loc[i].values[0]
    
    sequence = nc_01_input.index.values
    for i in sequence:
        table.loc[i, 'nc_01_input'] = nc_01_input.loc[i].values[0]
        
    sequence = nc_02_input.index.values
    for i in sequence:
        table.loc[i, 'nc_02_input'] = nc_02_input.loc[i].values[0]
        
    sequence = kd_01_input.index.values
    for i in sequence:
        table.loc[i, 'kd_01_input'] = kd_01_input.loc[i].values[0]
        
    sequence = kd_02_input.index.values
    for i in sequence:
        table.loc[i, 'kd_02_input'] = kd_02_input.loc[i].values[0]
    
    table = table.set_index(keys = 'read_number', drop=False)
    
    return table
```

```python
fgd5_result = make_table(fgd5)
emx2os_result = make_table(emx2os)
ac_result = make_table(ac)
rp11_result = make_table(rp11)
ctd_result = make_table(ctd)
jpx_result = make_table(jpx)
mapkapk_h3k4me3_result = make_table(mapkapk_h3k4me3)
mapkapk_h3k27me3_result = make_table(mapkapk_h3k27me3)
```

```python
def first_0(table_result):
    
    for i, elem in enumerate(table_result['nc_01']):
        if 0 == elem:
            nc_01 = i
            break
    for i, elem in enumerate(table_result['nc_02']):
        if 0 == elem:
            nc_02 = i
            break
    for i, elem in enumerate(table_result['kd_01']):
        if 0 == elem:
            kd_01 = i
            break
    for i, elem in enumerate(table_result['kd_02']):
        if 0 == elem:
            kd_02 = i
            break
    for i, elem in enumerate(table_result['nc_01_input']):
        if 0 == elem:
            nc_01_input = i
            break
    for i, elem in enumerate(table_result['nc_02_input']):
        if 0 == elem:
            nc_02_input = i
            break
    for i, elem in enumerate(table_result['kd_01_input']):
        if 0 == elem:
            kd_01_input = i
            break
    for i, elem in enumerate(table_result['kd_02_input']):
        if 0 == elem:
            kd_02_input = i
            break
    
    first = pd.DataFrame({"sample:":['nc_01', 'nc_02', 'kd_01', 'kd_02', 'nc_01_input', 'nc_02_input', 'kd_01_input', 'kd_02_input'], "0_reads_from:":[nc_01, nc_02, kd_01, kd_02, nc_01_input, nc_02_input, kd_01_input, kd_02_input]})
    
    return(first)
```
